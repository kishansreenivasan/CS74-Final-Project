```python
teams = pd.read_csv("data/game_scores.csv")
```

```python
teams = teams[~teams['Season'].isin(["2000/01","2001/02", "2002/03","2003/04","2004/05","2005/06","2006/07","2007/08","2008/09"
,"2009/10","2010/11","2011/12","2012/13","2013/14","2014/15","2015/16"])]
```

```python
teams.to_csv('data/game_scores.csv', index=False)
```

```python
# Create home team perspective
home_df = teams.copy()
home_df['Team'] = home_df['HomeTeam']
home_df['Opponent'] = home_df['AwayTeam']
home_df['IsHome'] = 1

# Rename columns for home team
home_df = home_df.rename(columns={
    'FullTimeHomeGoals': 'team_goals',
    'FullTimeAwayGoals': 'team_goals_allowed',
    'HomeShots': 'team_shots',
    'AwayShots': 'team_shots_allowed',
    'HomeShotsOnTarget': 'team_shots_on_target',
    'AwayShotsOnTarget': 'team_shots_on_target_allowed',
    'HomeCorners': 'team_corners',
    'AwayCorners': 'team_corners_allowed',
    'HomeFouls': 'team_fouls',
    'AwayFouls': 'team_fouls_allowed',
    'HomeYellowCards': 'team_yellow_cards',
    'AwayYellowCards': 'team_yellow_cards_allowed',
    'HomeRedCards': 'team_red_cards',
    'AwayRedCards': 'team_red_cards_allowed'
})

# Create away team perspective
away_df = teams.copy()
away_df['Team'] = away_df['AwayTeam']
away_df['Opponent'] = away_df['HomeTeam']
away_df['IsHome'] = 0

away_df = away_df.rename(columns={
    'FullTimeAwayGoals': 'team_goals',
    'FullTimeHomeGoals': 'team_goals_allowed',
    'AwayShots': 'team_shots',
    'HomeShots': 'team_shots_allowed',
    'AwayShotsOnTarget': 'team_shots_on_target',
    'HomeShotsOnTarget': 'team_shots_on_target_allowed',
    'AwayCorners': 'team_corners',
    'HomeCorners': 'team_corners_allowed',
    'AwayFouls': 'team_fouls',
    'HomeFouls': 'team_fouls_allowed',
    'AwayYellowCards': 'team_yellow_cards',
    'HomeYellowCards': 'team_yellow_cards_allowed',
    'AwayRedCards': 'team_red_cards',
    'HomeRedCards': 'team_red_cards_allowed'
})

# Create team_win column for both
home_df['team_win'] = (home_df['FullTimeResult'] == 'H').astype(int)
away_df['team_win'] = (away_df['FullTimeResult'] == 'A').astype(int)

# Select the columns you want to keep
cols_to_keep = ['Season', 'MatchDate', 'Team', 'Opponent', 'IsHome',
                'team_goals', 'team_goals_allowed', 'team_win',
                'team_shots', 'team_shots_allowed',
                'team_shots_on_target', 'team_shots_on_target_allowed',
                'team_corners', 'team_corners_allowed',
                'team_fouls', 'team_fouls_allowed',
                'team_yellow_cards', 'team_yellow_cards_allowed',
                'team_red_cards', 'team_red_cards_allowed']

# Combine both perspectives
team_df = pd.concat([home_df[cols_to_keep], away_df[cols_to_keep]], ignore_index=True)

# Sort by date
team_df = team_df.sort_values(['Season', 'MatchDate']).reset_index(drop=True)
```

```python
team_df.to_csv('data/game_scores.csv', index=False)
```
